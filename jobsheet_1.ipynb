```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/wbc.csv')
df.head()
```

```python
# shape of the data
df.shape
```

```python
# describe dataframe
df.describe()
```

```python
# get data type
df.dtypes
```

```python
df['diagnosis'] = df['diagnosis'].astype('category')
# Assigning numerical values and storing in another column
df['diagnosis'] = df['diagnosis'].cat.codes
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[df.select_dtypes(include='number').columns] = scaler.fit_transform(df[df.select_dtypes(include='number').columns])
```

```python
df.head()
```

```python
correlation_values = df.corr()['diagnosis'].drop('diagnosis')

# Rank the columns by their correlation values
correlation_ranking = correlation_values.abs().sort_values(ascending=False)
correlation_ranking
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Specify the features and the target variable
X = df[['concave points_worst', 'perimeter_worst']]
y = df['diagnosis']

# Perform a stratified 80:20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Display the shapes of the resulting sets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```
